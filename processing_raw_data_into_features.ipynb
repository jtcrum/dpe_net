Raw data format: [Framework, T-Site, O-Site, DPE, O-Site Rings, T-Site Rings, Framework Rings]

```python
import numpy as np
```

```python
# Load Data
d = np.loadtxt('raw_data_full_set.csv',delimiter = ',',dtype=object)
```

```python
# funtion to turn rings into a vector of features
def rings2features(r):
    r = r.split('-')
    f = np.zeros(20)
    for R in r:
        R = int(R)
        f[R]+=1
    return f
```

# Full Data Set

```python
# Save the full data set in directory: full_data

full_set_x = []
full_set_y = []
for line in d:
    temp = rings2features(line[4])
    temp = list(temp)
    temp2 = rings2features(line[5])
    temp2 = list(temp2)
    temp3 = rings2features(line[6])
    temp3 = list(temp3)
    for p in temp2:
        temp.append(p)
    for p in temp3:
        temp.append(p)
    full_set_x.append(temp)
    full_set_y.append([line[3]])

# Save X data as csv
full_set_x = np.array(full_set_x)
np.savetxt('full_data/x_values.csv',full_set_x,delimiter=',',fmt='%d')

# Save Y Data as csv
full_set_y = np.array(full_set_y,dtype=float)
np.savetxt('full_data/y_values.csv',full_set_y,delimiter=',',fmt='%f')
```

# 75% Training / 25% Validating
One proton location from each T Site is removed from the training data, and added to the validating data.\
This data was preseparated in processing_raw_data_into_features.ipynb.

```python
np.random.seed(1) # Set seed so this is reproducible
fws = ['BEA','CHA','FAU','FER','MFI','MOR']
xtrain = []
xtl = []
xvalidate = []
xvl = []
ytrain = []
yvalidate = []

ts = []
os = []
for f in fws:
    tmpts = []
    for line in d:
        if line[0] == f:
            tmpts.append(line[1])
    tmpts = np.array(tmpts)
    tmpts = np.unique(tmpts)
    ts.append(tmpts)
for i, f in enumerate(fws):
    on = []
    for j, t in enumerate(ts[i]):
        tmpos =[]
        for line in d:
            if line[0] ==f:
                if line[1] == t:
                    tmpos.append(line[2])
        on.append(len(tmpos))
    os.append(on)

for i, f in enumerate(fws):
    for j, t in enumerate(ts[i]):
        olines = []
        n = os[i][j]
        r = np.random.randint(n)
        for k, line in enumerate(d):
            if line[0]==f and line[1]==t:
                olines.append(k)
        line = d[olines[r]] 

        # label information
        temp = line[0:3]
        xvl.append(temp)

        # x data
        temp = rings2features(line[4])
        temp = list(temp)
        temp2 = rings2features(line[5])
        temp2 = list(temp2)
        temp3 = rings2features(line[6])
        temp3 = list(temp3)
        for p in temp2:
            temp.append(p)
        for p in temp3:
            temp.append(p)
        #                 for i in range(8,13):
        #                     temp.append(line[i])
        xvalidate.append(temp)

        # y data
        yvalidate.append([line[3]])
        
        del olines[r]
        
        for q in olines:
            line = d[q]
            # label information
            temp = line[0:3]
            xtl.append(temp)

            # x data
            temp = rings2features(line[4])
            temp = list(temp)
            temp2 = rings2features(line[5])
            temp2 = list(temp2)
            temp3 = rings2features(line[6])
            temp3 = list(temp3)
            for p in temp2:
                temp.append(p)
            for p in temp3:
                temp.append(p)
#                 for i in range(8,13):
#                     temp.append(line[i])
            xtrain.append(temp)

            #y data

            ytrain.append([line[3]])

# Save X training data as csv
xtrain = np.array(xtrain)
np.savetxt('one_per_T/x_train.csv',xtrain,delimiter=',',fmt='%d')

# Save Y training data as csv
ytrain = np.array(ytrain,dtype=float)
np.savetxt('one_per_T/y_train.csv',ytrain,delimiter=',',fmt='%f')

# Save X validating data as csv
xvalidate = np.array(xvalidate)
np.savetxt('one_per_T/x_validate.csv',xvalidate,delimiter=',',fmt='%d')

# Save Y validating data as csv
yvalidate = np.array(yvalidate,dtype=float)
np.savetxt('one_per_T/y_validate.csv',yvalidate,delimiter=',',fmt='%f')
```

# Random 70% Training / 30% Validating
Fully random splitting of the entire data set. 

```python
# Get x and y train/validate data in torch form
# ~70% of total data set
# ~30% of total data set

np.random.seed(1) # Set seed so this is reproducible
fws = ['BEA','CHA','FAU','FER','MFI','MOR']
xtrain = []
xtl = []
xvalidate = []
xvl = []
ytrain = []
yvalidate = []

for line in d:
    r = np.random.random()
    if r >= 0.3:

        # label information
        temp = line[0:3]
        xtl.append(temp)

        # x data
        temp = rings2features(line[4])
        temp = list(temp)
        temp2 = rings2features(line[5])
        temp2 = list(temp2)
        temp3 = rings2features(line[6])
        temp3 = list(temp3)
        for p in temp2:
            temp.append(p)
        for p in temp3:
            temp.append(p)
        xtrain.append(temp)

        #y data

        ytrain.append([line[3]])
    else:

        # label information
        temp = line[0:3]
        xvl.append(temp)

        # x data
        temp = rings2features(line[4])
        temp = list(temp)
        temp2 = rings2features(line[5])
        temp2 = list(temp2)
        temp3 = rings2features(line[6])
        temp3 = list(temp3)
        for p in temp2:
            temp.append(p)
        for p in temp3:
            temp.append(p)
        xvalidate.append(temp)

        # y data
        yvalidate.append([line[3]])

        
        
# Save X training data as csv
xtrain = np.array(xtrain)
np.savetxt('random_30/x_train.csv',xtrain,delimiter=',',fmt='%d')

# Save Y training data as csv
ytrain = np.array(ytrain,dtype=float)
np.savetxt('random_30/y_train.csv',ytrain,delimiter=',',fmt='%f')

# Save X validating data as csv
xvalidate = np.array(xvalidate)
np.savetxt('random_30/x_validate.csv',xvalidate,delimiter=',',fmt='%d')

# Save Y validating data as csv
yvalidate = np.array(yvalidate,dtype=float)
np.savetxt('random_30/y_validate.csv',yvalidate,delimiter=',',fmt='%f')
```
